```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://statiz.sporki.com/stats/?m=main&m2=fielding&m3=default&so=WAAwithPOS&ob=DESC&year=2023&sy=&ey=&te=&po=&lt=10100&reg=A&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=50&ph=&hs=&us=&na=&ls=&sf1=&sk1=&sv1=&sf2=&sk2=&sv2='
response = requests.get(url)
```

```python
html = response.text
soup = BeautifulSoup(html, 'html.parser')
temp = soup.find_all("table")[0]
```

```python
df = pd.DataFrame(index = range(50), columns = ["Rank", "Name", "Team", "WAAwithPOS", "G", "GS", "IP", "TC", "PO", "Ass", "E", "F%", "RF9", "Range RAA",
                                                 "Arm RAA", "Err RAA", "DP RAA", "Bt RAA", "CS RAA", "Blk RAA", "Frm RAA", "종합 RAA", "/144", "POSAdj", "RAAwithPOS", "WAAwoPOS", "WAAwithPOS"])

df
```

```python
l = 0
trs = temp.find_all("tr")
for j in range(2, min(414, len(trs))):  # 리스트의 길이를 벗어나지 않도록 조정
    temp2 = trs[j]
    if len(temp2.find_all("td")) == 27:
        for i in range(27):
            tds = temp2.find_all("td")
            if i < len(tds):  # 리스트의 길이를 벗어나지 않도록 조정
                temp3 = tds[i]
                df.iloc[l,i] = temp3.get_text()
            else:
                # 만약 인덱스가 범위를 벗어난 경우 처리할 내용 추가
                pass
        l += 1

df
```
