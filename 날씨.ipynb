- 날짜 형식 다름

```python
# 날짜 컬럼을 연도, 월, 일로 분리
weather_df[['연도', '월', '일']] = weather_df['일시'].str.split('-', expand=True).astype(int)
weather_df.drop(columns=['일시'], inplace=True)
```

```python
weather_df
```

```python
spec_df['지점명'] = spec_df['경기장'].str[:2]
```

```python
spec_df
```

```python
# 데이터프레임 병합
weather_spec = pd.merge(weather_df, spec_df, on=['연도', '월', '일', '지점명'])

print(weather_spec)
```

```python
weather_spec
```

```python
def discomfort_index(평균기온_x, 습도):
    """
    Calculate discomfort index (DI) given temperature in Celsius and relative humidity.
    
    Parameters:
        temperature_C (float): Temperature in Celsius.
        humidity (float): Relative humidity in percentage.
    
    Returns:
        float: Discomfort index.
    """
    DI = 0.81 * 평균기온_x + 0.01 * 습도 * (0.99 * 평균기온_x - 14.3) + 46.3
    return DI

# 불쾌지수 계산 후 새로운 컬럼에 추가
weather_spec['불쾌지수'] =round(discomfort_index(weather_spec['평균기온_x'], weather_spec['습도']))

print(weather_spec)
```
