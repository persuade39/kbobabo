```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://statiz.sporki.com/stats/?m=main&m2=batting&m3=default&so=WAR&ob=DESC&year=2022&sy=&ey=&te=&po=&lt=10100&reg=R&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=50&ph=&hs=&us=&na=&ls=&sf1=&sk1=&sv1=&sf2=&sk2=&sv2='

response = requests.get(url)
```

```python
html = response.text
soup = BeautifulSoup(html, 'html.parser')
temp = soup.find_all("table")[0]
```

```python
df = pd.DataFrame(index = range(50), columns = ["Rank", "Name", "Team", "WAR", "G", "PA", "ePA", "AB", "R", "H", "2B", "3B", "HR", "TB", 
                                                 "RBI", "SB", "CS", "BB", "HP", "IB", "SO", "GDP", "SH", "SF", "AVG", "OBP", 
                                                 "SLG", "OPS", "R/ePA", "wRC+", "WAR"])

df
```

```python
l = 0
trs = temp.find_all("tr")
for j in range(2, min(414, len(trs))):  # 리스트의 길이를 벗어나지 않도록 조정
    temp2 = trs[j]
    if len(temp2.find_all("td")) == 31:
        for i in range(31):
            tds = temp2.find_all("td")
            if i < len(tds):  # 리스트의 길이를 벗어나지 않도록 조정
                temp3 = tds[i]
                df.iloc[l,i] = temp3.get_text()
            else:
                # 만약 인덱스가 범위를 벗어난 경우 처리할 내용 추가
                pass
        l += 1

df
```
